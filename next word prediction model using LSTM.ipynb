```python
data = """YOUR_DATA"""
```

# converting textual data into numbers¶

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

#### Tokenizer: This is a class from the tensorflow.keras.preprocessing.text module used for text preprocessing.

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in data.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])  #slicing
```

```python
input_sequences
```

```python
max_length = max([len(x) for x in input_sequences]) # har ek seq ka len nikala fir usme se max dhundha
max_length
```

### sari sequences k shape ko same krne k liye


```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_length, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]  #saari ki saari rows but columns last vale ko chod k
```

```python
X
```

```python
 Y = padded_input_sequences[:,-1]  #: : Select all rows. -1: Select the last element of each row (since -1 is a shorthand for the last element in Python indexing).
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from tensorflow.keras.utils import to_categorical
Y= to_categorical(Y, num_classes=451)   #len(tokenizer.word_index)+1
```

```python
Y.shape
```

```python
Y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(451, 100, input_length=30))
model.add(LSTM(150))
model.add(Dense(451, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,Y,epochs=100)
```

```python
import time
import numpy as np
text = "As Tanisha progressed"

for i in range(10):
    token_text = tokenizer.texts_to_sequences([text])[0]  
    padded_token_text = pad_sequences([token_text], maxlen=30, padding='pre')
    
    pos = np.argmax(model.predict(padded_token_text))

    for word, index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            print(text)
            time.sleep(2)  
```
